```python
from nltk.book import text5
from nltk.probability import FreqDist
```

```python
four_letter_word = ([word.lower() for word in text5 if word.isalpha() and len(word) == 4])
print(four_letter_word)
```

```python
fdist = FreqDist(four_letter_word)
print(fdist)
```

```python
fdist.plot(25)
```
